***Image to pencil sketch***

Description of task: We need to read the image in RBG format and then convert it to a pencil sketch by using OpenCV library. .

Importing libraries

```python
#import libraries
import numpy as np
import cv2
%matplotlib inline 
import matplotlib.pyplot as plt
```

Reading Original Image

```python
image = cv2.imread("Humayuns-Tomb.jpg")
plt.imshow(image[:,:,::-1])
```

Converting original image to grayscale

```python
gray_image=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
plt.imshow(gray_image)
```

Inverting the grayscale image

```python
inverted_gray_image=255-gray_image
plt.imshow(inverted_gray_image)
```

Blurring the image using Gaussian function

```python
blurred_image=cv2.GaussianBlur(inverted_gray_image,(21,21),0)
plt.imshow(blurred_image)
```

Inverting the blurred image

```python
inverted_blurred_image=255-blurred_image
plt.imshow(inverted_blurred_image)
```

Pencil Sketch

```python
sketch_image=cv2.divide(gray_image,inverted_blurred_image, scale=256.0)
pencil_sketch=cv2.cvtColor(sketch_image, cv2.COLOR_BGR2RGB)
plt.imshow(pencil_sketch)
```
